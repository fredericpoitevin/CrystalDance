Two LLM models have been proposed and their corresponding real-space correlation function are:
$$
\tilde{\Gamma}_{1}(q) \propto (1+(\gamma_{1} q)^{2})^{-1} \leftrightarrow \Gamma_{1}(r) \propto r^{-1}e^{-r/\gamma_{1}}\\
\tilde{\Gamma}_{2}(q) \propto (1+(\gamma_{2} q)^{2})^{-2} \leftrightarrow \Gamma_{2}(r) \propto e^{-r/\gamma_{2}}
$$
It should be noted that the first LLM model tends to a phonon-like model in the $\gamma_{1} \to \infty$ limit:
$$
\tilde{\Gamma}(q) \propto  q^{-2} \leftrightarrow \Gamma(r) \propto r^{-1}
$$
We illustrate below. Both LLM model could yield very similar results provided their respective correlation lengths are tuned accordingly. The phonon-like model resembles a long-range LLM model if the part of the halo most close to the Bragg peak is masked.
